```python
import matplotlib.pyplot as plt
capacity = 400

water_consumption = [60, 50, 60, 80, 30, 10, 30]
average = sum(water_consumption) / len(water_consumption)

print(f"Tank Capacity: {capacity} liters")
print(f"Average usage per day: {average:.2f} liters")
print(f"Water consumption in a week: {sum(water_consumption)} liters")

print(f"Tank Capacity: {capacity} liters\n")

remaining_water = capacity
week = 1

while remaining_water > 0:

    print(f"\nWeek {week}")

    for i in range(len(water_consumption)):

        remaining_water -= water_consumption[i]

        print(
            f"Day {i+1} -> Used {water_consumption[i]} liters, "
            f"Remaining = {remaining_water} liters"
        )

        if remaining_water < 0:
            print(f"\nWater shortage occurs on Week {week} Day {i+1}")
            break

    if remaining_water < 0:
        break

    week += 1
plt.bar(range(len(water_consumption)), water_consumption)
plt.xlabel('Days')
plt.ylabel('Liters')
plt.title('Water Consumption per Day')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('/content/weather_forecast_data.csv')

avg_humidity = df['Humidity'].mean()
avg_cloud = df['Cloud_Cover'].mean()

days = []
predictions = []

for i, row in enumerate(df.tail(7).itertuples(), start=1):

    humidity = row.Humidity
    cloud_cover = row.Cloud_Cover

    if humidity > avg_humidity and cloud_cover > avg_cloud:
        pred = "Rain"
        value = 1
    else:
        pred = "No Rain"
        value = 0

    days.append(f"Day {i}")
    predictions.append(value)

plt.figure(figsize=(8,5))
plt.plot(days, predictions, marker='o')

plt.title("7-Day Weather Prediction Trend")
plt.xlabel("Days")
plt.ylabel("Prediction")
plt.yticks([0,1], ["No Rain", "Rain"])

plt.grid(True)
plt.show()
```
