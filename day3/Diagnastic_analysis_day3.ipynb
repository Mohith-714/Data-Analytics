```python
import pandas as pd
import numpy as np
customer_name=["A","B","C","D","E"]
internet_speed=[100,200,300,100,200]
payment_failure=[2,3,0,1,5]
delivery_charges=[200,300,250,400,350]
purchase_completed=["False","False","True","False","True"]
print("-----Dataset uploaded successfully-----")
for i in range(len(customer_name)):
  print(f"Customer Name: {customer_name[i]}, Internet Speed: {internet_speed[i]}, Payment Failure: {payment_failure[i]}, Delivery Charge: {delivery_charges[i]}, Purchase Completed: {purchase_completed[i]}")

print("-----Customers Aboundment-----")
total_aboundment = purchase_completed.count("False")
print(f"Total Aboundment (customers who did not complete purchase): {total_aboundment}")
for i in range(len(customer_name)):
  if purchase_completed[i] == "False":
    print(f"Customer Name: {customer_name[i]}, Internet Speed: {internet_speed[i]}, Payment Failure: {payment_failure[i]}, Delivery Charge: {delivery_charges[i]}")

print("-----Payment Diagnostic-----")
payment_diagnasic=sum(1 for failure in payment_failure if failure > 0)
print(f"Total Payment Failures (non-zero entries): {payment_diagnasic}")
for i in range(len(customer_name)):
  if payment_failure[i] > 0:
    print(f"Customer Name: {customer_name[i]}, Internet Speed: {internet_speed[i]}, Payment Failure: {payment_failure[i]}, Delivery Charge: {delivery_charges[i]}")

print("-----Internet Speed Diagnostic-----")
more_diagnatic_charges=sum(1 for speed in internet_speed if speed < 200)
print(f"Number of customers with internet speed < 200: {more_diagnatic_charges}") # Corrected description
for i in range(len(customer_name)):
  if internet_speed[i]<200:
    print(f"Customer Name: {customer_name[i]}, Internet Speed: {internet_speed[i]}, Payment Failure: {payment_failure[i]}, Delivery Charge: {delivery_charges[i]}")

print("-----Root Cause Analysis-----")
for i in range(len(customer_name)):
  if purchase_completed[i]=="False":
    if internet_speed[i]<200:
      print("internet_speed[i]<200")
      print(f"Customer Name: {customer_name[i]}, Internet Speed: {internet_speed[i]}, Payment Failure: {payment_failure[i]}, Delivery Charge: {delivery_charges[i]}")
    elif payment_failure[i]>0:
      print("payment_failure[i]>0")
      print(f"Customer Name: {customer_name[i]}, Internet Speed: {internet_speed[i]}, Payment Failure: {payment_failure[i]}, Delivery Charge: {delivery_charges[i]}")
    elif delivery_charges[i]>200:
      print("delivery_charges[i]>200")
      print(f"Customer Name: {customer_name[i]}, Internet Speed: {internet_speed[i]}, Payment Failure: {payment_failure[i]}, Delivery Charge: {delivery_charges[i]}")
import matplotlib.pyplot as plt
plt.bar(customer_name, internet_speed)
plt.xlabel("Customer Name")
plt.ylabel("Internet Speed")
plt.title("Internet Speed Analysis")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

df = pd.read_csv('/content/internet_service_churn.csv')

print("=== Dataset Head ===")
display(df.head())
total_abandonment = df['churn'].sum()
print(f"\nTotal abandonment (churned customers): {total_abandonment}")

print("\nCustomers with Low Internet Speed :")
low_internet_speed_customers = df[df['download_avg'] < 5]
print(f"Total customers with Low Internet Speed: {len(low_internet_speed_customers)}")

print("\nCustomers Facing High Service Failure (service_failure_count > 3):")
high_service_failure_customers = df[df['service_failure_count'] > 3]
print(f"Total customers facing High Service Failure: {len(high_service_failure_customers)}")

print("\nCustomers Facing High Bill Average (bill_avg > 30):")
high_bill_avg_customers = df[df['bill_avg'] > 30]
print(f"Total customers facing High Bill Average: {len(high_bill_avg_customers)}")


plt.plot(df['id'], df['download_avg'])
plt.xlabel("Customer ID")
plt.ylabel("Download Average")
plt.title("Customer Internet Speed (Download Average)")
plt.show()

plt.scatter(df['id'], df['service_failure_count'])
plt.xlabel("Customer ID")
plt.ylabel("Service Failure Count")
plt.title("Customer Service Failure Count")
plt.show()

plt.plot(df['id'], df['bill_avg'])
plt.xlabel("Customer ID")
plt.ylabel("Bill Average")
plt.title("Customer Bill Average")
plt.show()
```
