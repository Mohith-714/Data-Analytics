```python
patient_name=["A","B","C","D","E","F"]
age=[40,68,55,70,20,35]
patient_codition=["Good","Bad","Good","Bad","Bad","Good"]
oxygen_level=[110,80,90,85,120,120]
no_of_beds=4
count=0
for i in range(len(patient_name)):
  print(f"Patient Name:{patient_name[i]},Age:{age[i]},Patient Condition:{patient_codition[i]},Oxygen Level:{oxygen_level[i]}")
print("patient Analysis")
for i in range(len(patient_name)):
  if patient_codition[i]=="Bad":
    if age[i]>60 or oxygen_level[i]<90:
      print(f"Patient Name:{patient_name[i]},Age:{age[i]},Patient Condition:{patient_codition[i]},Oxygen Level:{oxygen_level[i]}")
      count+=1
      if count==no_of_beds:
        print("Required beds are not available")
        break
print(f"Total number of beds required:{count}")
import matplotlib .pyplot as plt
plt.bar(patient_name,oxygen_level)
plt.xlabel("Patient Name")
plt.ylabel("Oxygen Level")
plt.title("Oxygen Level Analysis")
plt.show()
```
