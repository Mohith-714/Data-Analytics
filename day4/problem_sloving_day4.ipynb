```python
products = ["Milk", "Bread", "Rice", "Oil"]
daily_sales = [5000, 2000, 1000, 4000]
product_stock = [20, 0, 15, 10]

# Fast-moving product (highest sales)
max_sales = max(daily_sales)
fast_moving = products[daily_sales.index(max_sales)]

# Slow-moving product (lowest sales)
min_sales = min(daily_sales)
slow_moving = products[daily_sales.index(min_sales)]

# Out-of-stock products
out_of_stock = []

for i in range(len(products)):
    if product_stock[i] == 0:
        out_of_stock.append(products[i])

print("Fast-moving product:", fast_moving)
print("Slow-moving product:", slow_moving)
print("Out-of-stock products:", out_of_stock)
import matplotlib.pyplot as plt
plt.pie(daily_sales,labels=products,autopct="%1.1f%%")
plt.show()
```
